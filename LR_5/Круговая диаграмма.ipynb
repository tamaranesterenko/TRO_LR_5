### Постройте круговую диаграмму населений стран мира

# Решение

```python
import matplotlib.pyplot as plt
%matplotlib inline

Country = ['China', 'India', 'USA', 'Indonesia', 'Pakistan', 'Nigeria',
         'Blazilia', 'Bangladesh', 'Russia']
Popular = [1452430137, 1422778000, 333449261, 277311900, 229864662, 218770596,
          209659740, 180313453, 146033315]

fig, ax = plt.subplots(figsize=(24, 14))
ax.pie(Popular, labels=Country, wedgeprops=dict(width=0.5))
```
