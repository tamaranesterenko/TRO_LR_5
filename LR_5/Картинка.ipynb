# Отображение выбранного изображения средствами библиотеки matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
import requests
from io import BytesIO

response = requests.get('https://kartinkin.net/uploads/posts/2022-02/1644878293_1-kartinkin-net-p-tyulpani-kartinki-1.jpg')
img = Image.open(BytesIO(response.content))
plt.imshow(img)
```
